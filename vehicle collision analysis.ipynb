```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime 
from dateutil.parser import parse
```

```python
# reading the data from csv file
df=pd.read_csv("Documents/version_vehicle_collision.csv")
df
```

```python
# to know about the data(data exploration)
df.info()
```

```python
# lambda function is a small anonymous funtion that can be defined without name don't want to define a seperate function
df["CRASH DATE"]=df["CRASH DATE"].apply(lambda x:parse(x).strftime('%d/%m/%y'))
df
```

```python
# filling missing values(cleaning the data)
df=df.fillna({"BOROUGH":"NA","CONTRIBUTING FACTOR VEHICLE 1":"NA","CONTRIBUTING FACTOR VEHICLE 2":"NA","VEHICLE TYPE CODE 1":"NA","VEHICLE TYPE CODE 2":"NA"})
df
```

```python
# from whole data set I'm using or intrested in these columns
df=df[['CRASH DATE','CRASH TIME','BOROUGH','NUMBER OF PERSONS KILLED','NUMBER OF PEDESTRIANS INJURED','NUMBER OF PEDESTRIANS KILLED','NUMBER OF CYCLIST INJURED','NUMBER OF CYCLIST KILLED','NUMBER OF MOTORIST INJURED','NUMBER OF MOTORIST KILLED', 'CONTRIBUTING FACTOR VEHICLE 1','CONTRIBUTING FACTOR VEHICLE 2','COLLISION_ID','VEHICLE TYPE CODE 1','VEHICLE TYPE CODE 2']]
df
```

```python
df
```

```python
# the data is cleaned with no null values
df.info()
```

```python
# converting the crash date and crash time into datetime format in order to use it
df["CRASH DATE"]=pd.to_datetime(df["CRASH DATE"])
df["CRASH TIME"]=pd.to_datetime(df["CRASH TIME"])
```

```python
# data is cleaned with no null values and consistent data types and ready for data analysis
df.info()
```

# DATA ANALYSIS

```python
# Extracting date from crash date
print(df["CRASH DATE"].dt.date)
```

```python
# extracting the month from crash date
df["Month"]=df["CRASH DATE"].dt.month
df
```

```python
# extracting year from crash date
df["YEAR"]=df["CRASH DATE"].dt.year
df
```

```python
# crash frequency by month
crash_count_by_month=df["Month"].value_counts().sort_index()
crash_count_by_month 
```

```python
df["CRASH DATE"].sort_values()
```

```python
# crash frequency by year
crash_frequency_year=df["YEAR"].value_counts().sort_index()
crash_frequency_year
```

```python
# extracting hour and minutes from crash time
df["HOUR"]=df["CRASH TIME"].dt.hour
df["Minutes"]=df["CRASH TIME"].dt.minute
df
```

```python
df["HOUR"].head(20)
```

```python
# crash frequency by hours
no_collision_hour=df["HOUR"].value_counts().sort_index()
no_collision_hour
```

```python
# crash frequency by minutes
no_collision_minute=df["Minutes"].value_counts().sort_index()
no_collision_minute
```

```python
#crash frequency by borough
crash_frequency_borough=df["BOROUGH"].value_counts().sort_index()
crash_frequency_borough
```

```python
df["CONTRIBUTING FACTOR VEHICLE 1"]
```

```python
# crash frequency by contributing factors
contributing_factors=df["CONTRIBUTING FACTOR VEHICLE 1"].value_counts().sort_values().tail(15)
print(contributing_factors)
```

```python
df["VEHICLE TYPE CODE 1"].head(40)
```

```python
# crash frequency by vehicle type code
crash_frequency_vehicle_type=df["VEHICLE TYPE CODE 1"].value_counts().sort_values().tail(15)
crash_frequency_vehicle_type
```
